```python
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("benchmark.csv")
df
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df.m, df.p, c=df.speedup_splay, cmap="RdBu_r", vmin=0, vmax=2)
plt.colorbar()
plt.title("Splay tree vs BST speedup")
plt.xlabel("number of frequently accessed elements")
plt.ylabel("query frequency")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df.m, df.p, c=df.speedup_arraycache, cmap="RdBu_r", vmin=0, vmax=2)
plt.colorbar()
plt.title("BST with cache vs BST speedup")
plt.xlabel("number of frequently accessed elements")
plt.ylabel("query frequency")
```
