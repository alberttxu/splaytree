```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("benchmark.csv")
df
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df.m, df.p, c=df.speedup_splay, cmap="RdBu_r", vmin=0, vmax=2)
plt.colorbar()
plt.title("Splay tree speedup")
plt.xlabel("m (number of more frequently accessed elements)")
plt.ylabel("p (access frequency)")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df.m, df.p, c=df.speedup_arraycache, cmap="RdBu_r", vmin=0, vmax=2)
plt.colorbar()
plt.title("BST with cache speedup")
plt.xlabel("m (number of more frequently accessed elements)")
plt.ylabel("p (access frequency)")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df.m, df.p, c=df.speedup_presplay, cmap="RdBu_r", vmin=0, vmax=2)
plt.colorbar()
plt.title("Pre-splay speedup")
plt.xlabel("m (number of more frequently accessed elements)")
plt.ylabel("p (access frequency)")
```
